```python
#Посчитайте частные производные функции, градиент в точке (1,2)
import numpy as np
from scipy.optimize import approx_fprime
def func (x):
    return 2 * x[0] ** 2 * x[1] ** 3 + 1 / x[0] + x[1] ** 2 * x[0] + 7
x = np.ones(2)
x[0],x[1] = (1,2)
eps = np.sqrt(np.finfo(float).eps)
approx_fprime([1,2],func,[eps,eps])
```
